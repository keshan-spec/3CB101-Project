```python
import pandas as pd
import pandas_datareader as web
import datetime as dt
from matplotlib import pyplot
from statsmodels.tsa.seasonal import seasonal_decompose
import statsmodels.tsa.stattools as ts
import statsmodels.api as sm
import seaborn as sns
```

```python
series = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv', header=0, index_col=0 )
# series.index = pd.to_datetime(series.index, format='%Y')
```

```python
df = series.copy()
```

```python
df.reset_index(inplace=True)
df['Month'] = pd.to_datetime(df['Month'])
df = df.set_index('Month')

df 
```

```python
result = seasonal_decompose(df, model='multiplicative')
result.resid.plot()
```

```python
result.seasonal.plot()
```

```python
result.trend.plot()
```

```python
crypto_currency = "BTC"
against_currency = "GBP"

# get the data from Yahoo finance for the crypto currency and a real world currency
start = dt.datetime(2021,11,1)
end = dt.datetime.now()
data = web.DataReader(f"{crypto_currency}-{against_currency}", 'yahoo', start, end)
```

```python
# For a Value Error, try the following: 
pd.options.mode.chained_assignment = None 

data_copy = data.copy()
data_copy.drop(['Adj Close'], axis=1, inplace=True)
```

```python
data_copy
```

```python
df = data_copy[["Close"]]
df.reset_index(inplace=True)

df['Open'] = pd.to_numeric(df['Close'])
# convert the column (it's a string) to datetime type
df["Date"] = pd.to_datetime(df['Date'])

df = df.set_index('Date')
```

```python
df.loc[df.index > "2020-04-01"]
```

```python
# Freq : 
result = seasonal_decompose(df, model='multiplicative', freq=12)
result.plot()
pyplot.show()
```

```python
X = df['Close'].values

result = ts.adfuller(X)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
	print('\t%s: %.3f' % (key, value))
```

```python
sm.graphics.tsa.plot_acf(df['Close'], lags=50)
```

```python
corr = data_copy.corr()

cmap = cmap = sns.diverging_palette(5, 250, as_cmap=True)
corr.style.background_gradient(cmap, axis=1)\
    .set_properties(**{'max-width': '100px', 'font-size': '15pt'})\
    .set_precision(2)

#Close price correlation matrix
k = 6  # number of variables for heatmap
cols = data_copy.corr().nlargest(k, 'Close')['Close'].index
cm = data_copy[cols].corr()
pyplot.figure(figsize=(10, 6))
sns.heatmap(cm, annot=True, cmap='viridis')

l = data_copy.columns.values
number_of_columns = 6
number_of_rows = len(l)-1/number_of_columns
pyplot.figure(figsize=(number_of_columns, 5*number_of_rows))
for i in range(0, len(l)):
    pyplot.subplot(number_of_rows + 1, number_of_columns, i+1)
    sns.set_style('whitegrid')
    sns.boxplot(data_copy[l[i]], color='green', orient='v')
    pyplot.tight_layout()

pyplot.figure(figsize=(2*number_of_columns, 5*number_of_rows))
for i in range(0, len(l)):
    try:
        pyplot.subplot(number_of_rows + 1, number_of_columns, i+1)
        sns.distplot(data_copy[l[i]], kde=True)
    except TypeError as e:
        print("Error in subplot", i, e)
```
